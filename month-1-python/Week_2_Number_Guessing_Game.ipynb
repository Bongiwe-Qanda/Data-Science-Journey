Number Guessing Game
A command-line game where the computer picks a random number
between 1-10 and the user has 3 attempts to guess it.

Built by: Bongi
Week 2 - Month 1

```python
import random
print("==========WELCOME TO MY NUMBER GUESSING GAME==========")
print()
def guessing():
  attempts = 0
  max_attempts = 3
  random_number = random.randint(1,10)
  guessed = False

  while attempts < max_attempts:
    guess = int(input("Guess any number from 1-10: "))
    attempts += 1
    remaining = max_attempts - attempts

    if guess == random_number:
      print("You have successfully guessed it!!")
      guessed = True
      break
    if guess < random_number:
      print("Guess higher")

    if guess > random_number:
      print("Guess lower")

    if attempts > 0:
      print(f"You have {remaining} attempts remaining.")

  if not guessed:
    print(f"The number you were meant to guess was {random_number}. Better Luck next time.")

  print()
  print("==========THANK YOU FOR PLAYING==========")


guessing()
```
