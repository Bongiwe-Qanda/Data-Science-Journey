A program that allows a user to add students with their marks for three subjects, calculate each student's average and assign them a grade(A,B,C.D,F), view a full class report showing every student's average, grade and pass/fail status, see who the top student is and what the class average is and save the full report to a text file.

Everything is accessed through a menu and the program handles invalid outputs.

```python

student_list = []

def add_student(name,mark_1,mark_2,mark_3):

  average = (mark_1+mark_2+mark_3)/3

  if average >= 90:
    grade  = "A"
  elif average >= 80:
    grade = "B"
  elif average >= 70:
    grade = "C"
  elif average >= 60:
    grade = "D"
  else:
    grade = "F"

  student_dict = {
      "name": name,
      "average" : average,
      "grade" : grade
  }

  student_list.append(student_dict)

  return student_list

def view_class_report():
  if len(student_list) == 0:
    print("No students in the class.")
  else:
    for student in student_list:
      print("-------------------------")
      print(f"Name: {student['name']}")
      print(f"Average: {student['average']}")
      print(f"Grade: {student['grade']}")
      print("--------------------------")

    class_summary()

def get_top_student():
    top = student_list[0]
    for student in student_list:
        if student["average"] > top["average"]:
            top = student
    return top


def class_summary():

  if len(student_list) == 0:
    print("No students in the class.")
  else:

    class_average = sum([student["average"] for student in student_list])/len(student_list)
    top_student = get_top_student()

    print(f"Class Average: {round(class_average, 2)}")
    print(f"Top Student: {top_student['name']} with {top_student['average']}")

def save_report_to_file():
  if len(student_list) == 0:
    print("No students in the class.")
  else:
    with open("class_report.txt", "w") as file:
      file.write("Class Report:\n")
      for student in student_list:
        file.write(f"Name: {student['name']}\n")
        file.write(f"Average: {student['average']}\n")
        file.write(f"Grade: {student['grade']}\n")
      print("Report saved to class_report.txt")


def menu():

  while True:
    print("\nMenu:")
    print("1. Add Student")
    print("2. View Class Report")
    print("3. Save report to file")
    print("4. Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
      name = input("Enter student's name: ")
      mark_1 = float(input("Enter student's mark for subject 1: "))
      mark_2 = float(input("Enter student's mark for subject 2: "))
      mark_3 = float(input("Enter student's mark for subject 3: "))
      add_student(name,mark_1,mark_2,mark_3)
      print("Student added successfully!")
    elif choice == "2":
      view_class_report()
    elif choice == "3":
      save_report_to_file()
    elif choice == "4":
      print("Goodbye!")
      break
    else:
      print("Invalid choice. Please try again.")
menu()


```
