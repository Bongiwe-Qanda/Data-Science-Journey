A program that calculates a user's BMI and tell them which category they fall  into.

BMI = weight / (height × height)

Below 18.5 → Underweight
18.5 to 24.9 → Normal weight
25 to 29.9 → Overweight
30 and above → Obese

```python
print("=======WELCOME TO THE BMI CALCULATOR======")
name = input("What is your name? ")
weight = float(input("What is your weight in kgs? "))
height = float(input("What is your height in meters? "))

bmi = round(weight / (height * height),2)

if bmi >= 30:
  print(f"Hello {name}, your BMI is {bmi} - you are obese.")
elif 25 <= bmi <= 29.9:
  print(f"Hello {name}, your BMI is {bmi} - you are overweight.")
elif 18.5 <= bmi <= 24.9:
  print(f"Hello {name}, your BMI is {bmi} - you have normal weight.")
else:
  print(f"Hello {name}, your BMI is {bmi} - you are underweight.")
print("============================================")
print("NB: BMI is just one health indicator.")
```
